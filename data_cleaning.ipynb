```python
df['employer_provider'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
df['employer_provider']
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df
```

### Salary Parsing

```python
df = df[df['Salary Estimate'] != '-1']
df
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
print(salary)
```

```python
minus_K_D = salary.apply(lambda x: x.replace('K','').replace('$',''))
minus_K_D
```

```python
min_hr = minus_K_D.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
min_hr.unique()
```

```python
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
df
```

```python
df['avg_salary'] = (df.min_salary+df.max_salary)/2
df
```

### Company Name text only

```python
df['company_text'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3],axis=1)
df
```

### State Field

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df
```

```python
df.job_state.value_counts()
```

```python
df['HC_same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
df
```

### Company Age


```python
df['age'] = df.Founded.apply(lambda x: x if x<1 else 2023-x)
df
```
